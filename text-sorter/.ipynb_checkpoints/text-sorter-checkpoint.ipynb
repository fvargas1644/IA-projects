# Preprocesamiento

***NLTK*** es una librería de Python que proporciona herramientas y recursos para trabajar con datos de lenguaje natural. Permite realizar tareas como análisis sintáctico, etiquetado de partes de la oración, análisis de sentimientos, tokenización, etc.

El método ***download()*** de NLTK se utiliza para descargar recursos y datasets necesarios para trabajar con la librería.

```python
nltk.download('all')
```

La función preprocess_text realiza cuatro pasos esenciales para preprocesar texto:

- ***text.lower():*** Convierte todo el texto a minúsculas. Esto es útil para asegurar que las palabras se traten de manera uniforme (sin diferenciar entre "Perro" y "perro").

- ***word_tokenize(text):*** Utiliza la función word_tokenize de la librería NLTK para dividir el texto en tokens (es decir, palabras o símbolos). Este proceso es conocido como tokenización.

- **Filtrado de tokens:**
    - ***word.isalnum():*** Asegura que solo se queden las palabras que son alfanuméricas, es decir, que no contengan caracteres especiales, puntuación, números, etc.
    - ***word not in stopwords.words('spanish'):*** Filtra las palabras que son stopwords (palabras vacías) en español, como "y", "de", "la", "el", etc. Las stopwords son palabras que generalmente no aportan valor significativo en el análisis de texto y se eliminan para que el análisis se enfoque en palabras más importantes.

- ***' '.join(tokens):*** Une los tokens filtrados nuevamente en una cadena de texto, separándolos por espacios. Es decir, genera una versión procesada del texto, pero con las palabras más relevantes (sin puntuación ni stopwords).

Finalmente, el DataFrame df se le agrega una nueva columna **'texto_procesado'**, que contiene los textos preprocesados, es decir, con los tokens relevantes y filtrados (en minúsculas, sin stopwords y sin caracteres no alfanuméricos).

```python
def preprocess_text(text):
    tokens = word_tokenize(text.lower())
    tokens = [word for word in tokens if word.isalnum() and word not in stopwords.words('spanish')]
    return ' '.join(tokens)

df['texto_procesado'] = df['textos'].apply(preprocess_text)
df
```

# Extracción de características y vectorización

Para transformar el texto depurado en características numéricas que los modelos puedan interpretar, utilizaremos dos técnicas: TF-IDF y Word2Vec.

- **TF-IDF (Term Frequency-Inverse Document Frequency):** Es una técnica utilizada en procesamiento de lenguaje natural (PLN) que valora la importancia de una palabra en un documento dentro de un conjunto de documentos. Se basa en dos componentes ***Frecuencia de término (TF)*** y ***Frecuencia inversa de documentos (IDF)***.

- **Word2Vec:** Es un modelo de aprendizaje profundo que genera representaciones vectoriales (embeddings) de palabras. Cada palabra se mapea a un vector en un espacio continuo de alta dimensión, donde las palabras con significados similares estarán representadas por vectores cercanos.

# Construcción de PIPELINE 

```python
# Cargar datos
df = pd.read_excel('dataset.xlsx')

spanish_stopwords = stopwords.words('spanish')

# Dividir datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(df['textos'], df['ODS'], test_size=0.2, random_state=42)

# Construcción del pipeline
pipeline = Pipeline([
    ('tfidf', TfidfVectorizer(stop_words=spanish_stopwords)),  # Representación vectorial con TF-IDF
    ('svd', PCA(n_components=100)),  # Reducción de dimensionalidad
    ('classifier', SVC(kernel='linear', C=1.0))  # Modelo de clasificación con SVM
])

# Entrenar el modelo
pipeline.fit(X_train, y_train)

# Evaluar el modelo
y_pred = pipeline.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)*100

print(f"L exactitud del modelo Support Vector Machine - TF IDF es de: %{round(accuracy, 2)}")
print(f"Matriz: {y_pred}")
```
